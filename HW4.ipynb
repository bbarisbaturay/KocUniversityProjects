```python
import pandas as pd
import numpy as np
```

```python
srv = pd.read_csv('immSurvey.csv')
srv.head()
```

```python
alphas = srv.stanMeansNewSysPooled
sample = srv.textToSend
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vec = CountVectorizer()
X = vec.fit_transform(sample)
X
```

```python
pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vec = TfidfVectorizer()
X = vec.fit_transform(sample)
pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, alphas, random_state=1)
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import ConstantKernel, RBF
```

```python
rbf = ConstantKernel(1.0)*RBF(length_scale=1.0)
gbr = GaussianProcessRegressor(kernel=rbf, alpha=1e-8)
gbr.fit(Xtrain.toarray(), ytrain)
```

```python
mu_s, cov_s = gbr.predict(Xtest.toarray(), return_cov=True)
```

```python
np.corrcoef(ytest, mu_s)
```

Correlation is 0.6833

BIGRAMS

```python
bigram_vectorizer = CountVectorizer(ngram_range=(1, 2), token_pattern=r'\b\w+\b')
```

```python
X = bigram_vectorizer.fit_transform(sample)
```

```python
pd.DataFrame(X.toarray(), columns=bigram_vectorizer.get_feature_names())
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, alphas, random_state=1)
```

```python
rbf = ConstantKernel(1.0)*RBF(length_scale=1.0)
gpr = GaussianProcessRegressor(kernel=rbf, alpha=1e-8)
gpr.fit(Xtrain.toarray(), ytrain)
```

```python
mu_s, cov_s = gpr.predict(Xtest.toarray(), return_cov=True)
np.corrcoef(ytest, mu_s)
```

Correlation became 0.6088 when bigrams are included. It was 0.6833 without bigrams.
